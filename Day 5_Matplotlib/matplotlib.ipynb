### Line plotting

For line plotting we need to create variables for both axises X and Y, as our dependant and independant variables, so we use linspace for creating a range of values for our X axis and a function to produce Y values.

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
y = np.sin(x)
# print(x)
# print(y)
ax.plot(x,y)
```

```python
# OBJECT-ORIENTED INTERFACE
fig=plt.figure()
ax=plt.axes()
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
```

```python
# MATLAB-STYLE INTERFACE
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
# ADJUSTING PLOT:LINE COLOR
plt.plot(x,np.sin(x),color='black')         # specify color by name
plt.plot(x,np.sin(x-1),color='g')          # short color code (rgbcymk)
plt.plot(x,np.sin(x-2),color='0.75')       # grayscale between 0 and 1
plt.plot(x,np.sin(x-3),color='#FFDD44')    # hex code
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3))# RGB tuple,values between 0 and 1
plt.plot(x,np.sin(x-5),color='chartreuse') # all HTML color names supported
```

```python
# ADJUSTING PLOT:LINE STYLE
plt.plot(x,x+0,linestyle='solid')
plt.plot(x,x+1,linestyle='dashed')
plt.plot(x,x+2,linestyle='dashdot')
plt.plot(x,x+3,linestyle='dotted')
#SHORTHAND NOTATIONS
plt.plot(x,x+4,linestyle='-')
plt.plot(x,x+5,linestyle='--')
plt.plot(x,x+6,linestyle='-.')
plt.plot(x,x+7,linestyle=':',color='black')
plt.show()
```

```python
#ADJUSTING THE PLOT:AXES LIMIT
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
#ADJUSTING THE PLOT:AXES LIMIT (ALTERNATIVE)
# PLT.AXIS METHOD ALLOWS U TO SET THE X AND Y LIMITS WITH A SINGLE
# CALL,BY PASSING A LIST THAT SPECIFIES[XMIN,XMAX,YMIN,YMAX]
plt.plot(x,np.sin(x))
plt.axis([2,11,-1.5,1.5])
```

```python
#PLT.AXIS METHOD ALLOWS AUTOMATICALLY TIGHTEN BOUNDS AROUND CURRENT
# PLOT
plt.plot(x,np.sin(x))
plt.axis('tight')
```

```python
#ENSURING EQUAL ASPECT RATIO SO THAT ON UR SCREEN 1 UNIT IN X = 1
# UNIT IN Y
plt.plot(x,np.sin(x))
plt.axis('equal')
```

```python
# LABELLING PLOT
plt.plot(x,np.sin(x))
plt.title('A sine curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# PLOT LEGEND EXAMPLE
plt.plot(x,np.sin(x),label='sin(x)')
plt.plot(x,np.cos(x),label='cos(x)')
plt.axis('equal')
plt.xlabel('x')
plt.legend()
```

##### LEGEND FOR SIZE OF POINTS (LATER)

```python
# MULTIPLE LEGENDS
fig,ax=plt.subplots()
lines=[]
styles=['-','--','-.',':']
x=np.linspace(0,10,1000)
for i in range(4):
    lines=lines+ax.plot(x,np.sin(x-i*np.pi/2),styles[i],color='k')
ax.axis('equal')
# specify lines and labels of first legend
plt.legend(lines[:2],['lineA','lineB'],loc='upper left')

# create the second legend and add the artist manually
from matplotlib.legend import Legend
leg=Legend(ax,lines[2:],['lineC','LineD'],loc='lower right',frameon=False)

ax.add_artist(leg)

# note if u try to create a second legend using plt.legend or ax.legend
# it will simply override the first one
    
```

### Pie Charts
Pie chart shows distribution of data based on proportion of pie occupied. 

```python
t_mov= pd.read_csv('datasets/types_movies.csv')
t_mov
```

```python
plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'])

plt.axis('equal')

plt.show()
```

#### Pie chart customizations
 We set the values for the colors and autopct properties. The latter sets the format for the values to be displayed. 

```python
colors = ['darkorange', 'sandybrown', 'darksalmon', 'orangered','chocolate']

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f')

plt.axis('equal')

plt.show()
```

#### The explode property
To highlight a particular wedge of the pie chart, we use explode to separate it from the rest of the chart. 

The value for "explode" represents the fraction of the radius with which to offset each wedge.

```python
explode = (0, 0.1, 0, 0, 0.3)

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f',
        explode=explode)

plt.axis('equal')

plt.show()
```

### Box Plot

```python
exam_data = pd.read_csv('datasets/exams.csv')
exam_data.head()
```

```python
exam_scores = exam_data[['math score', 'reading score', 'writing score']]
exam_scores.head()
```

```python
exam_scores.describe()
```

```python
# For plotting in boxplot, we convert it to an array
exam_scores_array = np.array(exam_scores)
exam_scores_array
```

```python
colors = ['blue', 'grey', 'lawngreen']

bp = plt.boxplot(exam_scores_array, 
                 patch_artist=True,
                 notch=True)

for i in range(len(bp['boxes'])):
    
    bp['boxes'][i].set(facecolor=colors[i])
    
    bp['caps'][2*i + 1].set(color=colors[i])
    
plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])

plt.show()
```

### Violin Plots
Similar to boxplots, except they can show the density of the data points around a particular value with their widths

```python
vp = plt.violinplot(exam_scores_array,
                    showmedians=True)

plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i])

plt.show()  
```

### Stack Plots
It is used stack plots and visualize data overtime.

```python
# Labels for each stack
labels = ['Badlands', 
          'GrandCanyon', 
          'BryceCanyon']

# Colors for each stack
colors = ['sandybrown', 
          'tomato', 
          'skyblue']

#  Similar to pandas df.plot.area()
plt.stackplot(x, y, 
              labels=labels,
              colors=colors, 
              edgecolor='black')

# Plots legend to the upperleft of Figure
plt.legend(loc=2)

plt.show()
```
